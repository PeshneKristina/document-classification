# Задание 2

нужно написать функцию, которая принимает две строки и проверяет, входит ли хотя бы одна перестановка второй строки в первую. Например:

> a = 'abcrotm'
> 
> b = 'tro'

функция def check_inv(a, b) вернет True, так как 'rot' содержится в 'abcrotm'. Нужно подумать как можно более оптимальный алгоритм и оценить его сложность. 

```python
def create_char_dict(string):
    char_dict = {}
    for i in range(0, len(string)):
        if char_dict.get(string[i]) is None:
            char_dict.update({string[i]: 1})
        else:
            char_dict[string[i]] = char_dict.get(string[i]) + 1
    return char_dict

def update_dict(dict,char):
    dict[char] = dict.get(char) - 1
    if dict[char] == 0:
        dict.pop(char)
    return dict
```

```python
import copy
def check_inv(given_string, search_string):
    dict_of_search_string = create_char_dict(search_string)
    copy_dict_of_search_string = copy.copy(dict_of_search_string)
    for i in range(0, len(given_string)):
        char = given_string[i]
        if copy_dict_of_search_string.get(char):
            copy_dict_of_search_string = update_dict(copy_dict_of_search_string,char)
            if len(copy_dict_of_search_string) == 0:
                return True
        else:
            copy_dict_of_search_string = copy.copy(dict_of_search_string)
    n_search_string = len(search_string)
    n_given_string = len(given_string)
    for j in range(0, n_search_string):
        char = given_string[n_given_string-1-j]
        if not dict_of_search_string.get(char):
            return False
        else:
            dict_of_search_string = update_dict(dict_of_search_string, char)
    return True
```

```python
a = 'abcrotm'
b = 'tro'
check_inv(a, b)==True
```

# Задание 3

Реализовать бинарное дерево (класс Tree), в нём методы __repr__, __str__, __iter__ (итерация только по листьям).

```python
class Tree:
    def __init__(self, value=None, left=None, right=None):
        self.left = left
        self.right = right
        self.value = value
        
    def __iter__(self):
        leaves = []
        if not self.left and not self.right:
            leaves.append(self.value)
        if self.left:
            l = self.left.__iter__()
            for el in l:
                leaves.append(el)
        if self.right:
            r = self.right.__iter__()
            for el in r:
                leaves.append(el)
        return iter(leaves)
    
    def __str__(self):
        s = ""
        if self.left:
            s += str(self.left).replace("\n", "\n   ")
        if self.left:
            s += "\n " + "/ "
        s += "\n" + str(self.value)
        if self.right:
            s += "\n " + "\\ "
        if self.right:
            s += str(self.right).replace("\n", "\n   ")
        return s
    
    def __repr__(self):
        if self.right and self.left:
            return '{}{}'.format(self.__class__.__name__,
                         repr((self.value, self.left, self.right)))
        elif self.left and not self.right:
            return '{}{}'.format(self.__class__.__name__,
                         repr((self.value, self.left)))
        elif self.right and not self.left:
            return '{}{}'.format(self.__class__.__name__,
                         repr((self.value, self.left)))
        else:
            return '{}({})'.format(self.__class__.__name__, repr(self.value))


```

```python
tree = Tree(0, Tree(1, Tree(3), Tree(4)),Tree(2))
list(tree) == [3, 4, 2]
```

# Задание 4

Реализовать простейший калькулятор математических выражений:
- только целые числа
- **+**, **\-**, **\***, **\**
- скобки

**Можно использовать регулярные выражения**

```python
import numbers

OPERATORS = {
    '+': (1, lambda x, y: x + y),
    '-': (1, lambda x, y: x - y),
    '*': (2, lambda x, y: x * y),
    '**': (2, lambda x, y: x ** y)
}


def get_list_expr(expr):
    list_expr = []
    number = ''
    for s in expr:
        if s in '1234567890.':
            number += s
        elif number:
            list_expr.append(int(number))
            number = ''
        if s in OPERATORS or s in "()":
            list_expr.append(s)
    if number:
        list_expr.append(int(number))
    return list_expr


def get_sorted_list_expr(list_expr):
    stack = []
    exit = []
    for s in list_expr:
        if isinstance(s, numbers.Number):
            exit.append(s)
        if s in OPERATORS:
            while stack and stack[-1] != "(" and OPERATORS[s][0] <= OPERATORS[stack[-1]][0]:
                op = stack.pop()
                exit.append(op)
            stack.append(s)
        if s == "(":
            stack.append(s)
        if s == ")":
            op = stack.pop()
            while op != "(":
                exit.append(op)
                op = stack.pop()
    while stack:
        op = stack.pop()
        exit.append(op)
    return exit
```

```python
def calc(expr):
    list_expr = get_list_expr(expr)
    sorted_list_expr = get_sorted_list_expr(list_expr)
    stack = []
    for el in sorted_list_expr:
        if isinstance(el, numbers.Number):
            stack.append(el)
        else:
            a = stack.pop()
            b = stack.pop()
            stack.append(OPERATORS[el][1](b, a))
    return stack.pop()
```

```python
calc('2 * (15 - 3 * 4) - 2') == 4
```
